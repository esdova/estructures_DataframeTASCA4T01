```python
import matplotlib.pyplot as plt                          # ok ex1 1.1
import pandas as pd

lloguer_df = pd.read_csv('preu_tritri.csv')
k=lloguer_df.groupby(["Nom_Districte"])["Nom_Districte"].count().plot(kind="bar") # agafo la columna Nom_districte la agrupo 
for container in k.containers:                                                    # per Nom_districte(variable categòrica) 
                                                                         # entenc, que cada districte seria una categoria) 
                                                                                  # a la gráfica de barres
    k.bar_label(container)# escric la quqntitat de registres que hi  
lloguer_df["Nom_Districte"].count()    #comprobo n'hi ha 584                      # ha de cada Nom_districte(categoria)
plt.savefig('graficaEX1_1.png')
```

#### Una variable númerica Preu

```python
import matplotlib.pyplot as plt
import pandas as pd                                   #EX 1.2 

lloguer_df = pd.read_csv('preu_tritri.csv')      # primer separo les files que mo tenen valor numeric
lloguer_df= lloguer_df[lloguer_df.Preu!="--" ]   #df_lloguer_mig es un nou dataframe té les files 
                                                 #de lloguer mitjà mensual(euro/mes)

df_lloguer_mig=lloguer_df.loc[lloguer_df['Lloguer_mitja'] == "Lloguer mitjà mensual (Euros/mes)"].loc[lloguer_df['Preu'] != "--"]
Preu=df_lloguer_mig['Preu'].astype("float")      #canvio el tipus de dades de la coluna preu  a float                        
Preu.plot() 
plt.xlabel("Preu mitjà (Euros/mes)")
plt.ylabel("Vegades")
plt.title("Preus X cops que surt aquest preu")
plt.legend()

plt.show()

plt.savefig('graficaEX1_2.png')
quants=Preu.count() #  numero de registres que compleixen que la columna LLoguer_mitja és "Lloguer mitjà mensual (Euros/mes)"
# i que no hi han valors "--" són 270

# comprobació
hiha =  sum (df_lloguer_mig['Preu'].astype("float")>=1800)      #comprobació vull saber si és veritat si hi ha un registre > 1800, 
quin_valor  = max(df_lloguer_mig['Preu'].astype("float"))       #i sí, hiha 1, li demano que em digui quin_valor té

quants,hiha,quin_valor
```

#### una variable numérica i una categórica(Districte i Preu)

```python
 #Exercici 1.3 OK
import matplotlib.pyplot as plt

lloguer_df = pd.read_csv('preu_tritri.csv')      # primer separo les files que mo tenen valor numeric
#lloguer_df= lloguer_df[lloguer_df.Preu!="--"] 

df_lloguer_mig=lloguer_df.loc[lloguer_df['Lloguer_mitja'] == "Lloguer mitjà mensual (Euros/mes)"].loc[lloguer_df['Preu'] != "--"]
df_lloguer_mig['Preu'] = df_lloguer_mig['Preu'].astype("float")
df_final = df_lloguer_mig.groupby(['Nom_Districte']).mean() # agrupo els nom_districte, mitja
df_final
df_final['Preu'].plot(kind='bar', title='Lloguer mitjà mensual (Euros/mes) per Districte') 

plt.xlabel( "Nom_Districtes")
plt.ylabel("Preu mitjà")
plt.title("Preu mitjà (Euros/mes)per Nom_Districte")
plt.legend()
plt.savefig('graficaEX1_3.png') #exporto la imatge de al gráfica 
```

#### Dos variables Numeriques Any i Preu

```python
#Exercici 1.4 OK

df_final14 = df_lloguer_mig.groupby(['Trimestre']).mean() # agrupo per trimestre i mostro per preu mig
df_final14
df_final14[('Preu')].astype(float).plot(kind='bar') #  preu  li canvio el tipus a float


plt.ylabel("Preu mig")
plt.title("Preu mig per trimestre")
plt.legend()
plt.savefig('graficaEX1_4.png')#exporto la imatge de al gráfica 
```

#### Tres variables nom districte, timestre i preu

```python
#Exercici 1.5 OK

df_final15 = df_lloguer_mig.groupby(['Nom_Districte', 'Trimestre']).mean() # agrupo Nom Districte y trimestre
df_final15['Preu'].plot(kind='bar', title='Lloguer mitjà mensual (Euros/mes) per Districte i Trimestre') # mostro per preu
plt.xlabel( "Nom_Districtes")
plt.ylabel("Preu mitjà")
plt.title("Preu mitjà (Euros/mes)per Nom_Districte i Trimestre")
plt.legend()
plt.savefig('graficaEX1_5.png')
```

```python

# EX 3 proposo visualitzar  els lloguers mes grans de 1000 dels diferents barris
#el més car és pedralbes

df_lloguer_mig_alt = lloguer_df.loc[lloguer_df['Lloguer_mitja'] == "Lloguer mitjà mensual (Euros/mes)"].loc[lloguer_df['Preu'] != "--"]
df_lloguer_mig_alt["Preu"] = df_lloguer_mig_alt["Preu"].astype("float")
df_lloguer_mig_alt=df_lloguer_mig_alt.loc[df_lloguer_mig_alt['Preu'] > 1000]

df_final3 = df_lloguer_mig_alt.groupby(['Nom_Barri']).mean()
df_final3['Preu'].plot(kind='bar', title='Lloguer mitjà mensual (Euros/mes) per Barri > 1000€ de mitja')

plt.xlabel( "Barris")
plt.ylabel("Preus")

plt.legend()
plt.savefig('graficaEX3.png')#exporto la imatge de al gráfica 
```
